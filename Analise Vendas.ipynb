```python
#Convertendo as de String para data e ajustando os formatos
def converter_data(data_str):
  if pd.isna(data_str):
    return pd.NaT
  formato_datas = [
      '%Y-%m-%d',
      '%d/%m/%Y',
      '%Y-%b-%d',
      '%Y.%m.%d',
      '%d-%m-%Y',
      '%Y/%m/%d',
      '%d/%m/%Y',
      '%d-%m-%Y',
  ]

  for formato in formato_datas:
    try:
      return pd.to_datetime(data_str, format=formato)
    except:
      continue
  try:
        return pd.to_datetime(data_str, dayfirst=True, errors='coerce')
  except:
        return pd.NaT
```

```python
df['data_ajustado'] = df['data_pedido'].apply(converter_data)
```

```python
#Verificando se existe algum valor errado na coluna antes de passar valores para float
df['produto_preco'].unique()
```

```python
#Passando os valores de String para Float
df['preco_str'] = df['produto_preco'].astype(str)

df['preco_str'] = df['preco_str'].str.replace('R$', '')
df['preco_str'] = df['preco_str'].str.replace('.', '', regex=False)
df['preco_str'] = df['preco_str'].str.replace(',', '.')
df['preco_str'] = df['preco_str'].str.strip()


df['produto_preco'] = pd.to_numeric(df['preco_str'], errors='coerce')


df.drop('preco_str', axis=1, inplace=True)
```

```python
df.info()
```

```python
#Verificando se possui algum erro de digitação em quantidade
df['quantidade'].unique()
```

```python
#Ajustando as quantidades que estão negativas
df.loc[df['quantidade']<0 ,'quantidade'] = abs(df.loc[df['quantidade']<0,'quantidade'])
```

```python
#Adicionando Uma nova coluna util para futuro dashboard
df['preco_total'] = df['produto_preco'] * df['quantidade']
```

```python
#Alterando o . por , para não dar erro no Power BI
df['produto_preco'] = df['produto_preco'].apply(lambda x: f"{x:.2f}".replace('.', ','))
```

```python
df['quantidade'] = df['quantidade'].apply(lambda x: f"{x:.2f}".replace('.', ','))
df['preco_total'] = df['preco_total'].apply(lambda x: f"{x:.2f}".replace('.', ','))
#Salvar arquivo limpo para dashboard
df.to_csv('dados_tratados.csv', index=False, encoding='utf-8-sig')
```
